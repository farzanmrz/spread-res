# GCP Environment Checker

We are going to check the variables setup for our GCP VM

```python
# Check the threads
import os
import subprocess

# Print the number of CPU threads
print(f"Number of CPU threads: {os.cpu_count()}")

# Check if CUDA is available
import torch
if torch.cuda.is_available():
    print("CUDA Available: True")

    # Capture and print the `nvidia-smi` output
    try:
        nvidia_smi_output = subprocess.check_output(["nvidia-smi"], text=True)
        print("\nNVIDIA-SMI Output:")
        print("-" * 80)
        print(nvidia_smi_output.strip())
        print("-" * 80)
    except FileNotFoundError:
        print("nvidia-smi not found. Please ensure NVIDIA tools are installed and accessible.")
else:
    print("CUDA Available: False")
```
